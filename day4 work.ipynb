for industry 52(Finance and Insurance) mean is 0.083234 stv is 0.03969
for industry 53(Real Estate and Rental and Leasing) mean is 0.089655  stv is 0.04184

```python
freq = (0.083234+0.089655)/8
variance_ = 8*(freq*(1-freq))/16
difference_ = stats.norm(0, np.sqrt(variance_))
```

```python
fig,ax=plt.subplots(1, figsize=(15,3))
x = np.linspace(-1,1, num=250)
ax.plot(x,difference_.pdf(x),linewidth=3)
ax.set_xlim(-1,1)
ax.set_title('null')
```
