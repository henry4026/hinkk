```python
def welch_test_statistic(finance_arr1_, real_estate_arr_):
    mean_ = np.mean(finance_arr1_) - np.mean(real_estate_arr_)
    sq = (np.var(finance_arr1_)/len(finance_arr1_)) + (np.var(real_estate_arr_)/len(real_estate_arr_))
    return mean_/np.sqrt(sq)
```

```python
welch_ = welch_test_statistic(finance_arr1, real_estate_arr)
welch_test_statistic(finance_arr1, real_estate_arr)
```

```python
def welch_satterhaithe_df(sample1, sample2):
    len_fin = len(sample1)
    len_rea = len(sample2)
    df = (
        ((np.var(sample1)/len_fin + np.var(sample2)/len_rea)**(2.0))/
        ((np.var(sample1)/len_fin)**(2.0)/(len_fin-1)+(np.var(sample2)/len_rea)**(2.0)/(len_rea-1)))
    return df
```

```python
df_welch = welch_satterhaithe_df(finance_arr1, real_estate_arr)
df_welch
```

```python
x = np.linspace(-3,3,num=250)
fig, ax = plt.subplots(1, figsize=(16,3))
turnover = stats.t(df_welch)
ax.plot(x, turnover.pdf(x), linewidth=2, label="degree of fredom: {welch}".format(welch = df_welch))
_ = ax.fill_between(x, turnover.pdf(x), where=(x >= -welch_), color='r', alpha=0.25)
_ = ax.fill_between(x, turnover.pdf(x), where=(x <= welch_), color='r', alpha=0.25)
ax.legend()
ax.set_title('p-value')
```

```python
p_value = turnover.cdf(welch_) + (1-turnover.cdf(-welch_))
p_value
```

```python
def count_winning_pairs(sample1, sample2):
    sample_1, sample_2 = np.array(sample1), np.array(sample2)
    n_total_wins = 0
    for x in sample1:
        n_wins = np.sum(x > sample2) + 0.5*np.sum(x == sample2)
        n_total_wins += n_wins
    return n_total_wins
```

```python
fin_win = count_winning_pairs(finance_arr1, real_estate_arr)
rea_win = count_winning_pairs(real_estate_arr,finance_arr1)
print((fin_win+rea_win),(len(finance_arr1)*len(real_estate_arr)))
```
